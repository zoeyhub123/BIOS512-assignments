Make a bar chart ðŸ“Š **sorted by frequency** of turbine count per province territory. (Hint: `fct_infreq` will sort your factor levels by frequency for subsequent plotting)

```r
options(repr.plot.height=6)

TidyTuesday|>
    mutate(province_territory=fct_infreq(TidyTuesday$province_territory))|>
    ggplot(aes(y=province_territory))+
    geom_bar()+
    labs(title = "Turbine Count per Province territory",
       x = "Province territory",
       y = "Turbine Count")
```

Make a bar chart **sorted by frequency** of turbine count per manufacturer. **Group the low frequency manufacturers into an `Other` category**.

**Bonus:** Can you force the `Other` category to be the last bar? 

```r
# check the number of manufacturers to find the possible threshold
TidyTuesday|>
  count(TidyTuesday$manufacturer, sort = TRUE)
```

```r
TidyTuesday|>
  count(manufacturer)|>                           
  mutate(manufacturer = if_else(n < 41, "Other", manufacturer))|>
  ggplot(aes(x = reorder(manufacturer, n), y = n)) +    # Reorder 
  geom_col() +
  coord_flip() +
  labs(title = "Wind Turbines by Manufacturer",
       x = "Manufacturer",
       y = "Number of Turbines")
```
